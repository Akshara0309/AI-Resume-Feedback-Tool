```python
import pandas as pd

resume = """
Name: Akshara

Skills:
Python
Pandas
Machine Learning
Data Analysis

Education:
B.E Computer Science

Projects:
Titanic Survival Analysis
House Price Prediction
"""

print("========== RESUME ==========\n")

print(resume)

required_skills = [
    "Python",
    "Pandas",
    "Machine Learning",
    "SQL",
    "Communication"
]

# Find Missing Skills
missing_skills = []

for skill in required_skills:
    if skill not in resume:
        missing_skills.append(skill)

# Calculate Score
score = (
    (len(required_skills) - len(missing_skills))
    / len(required_skills)
) * 100

# Display Feedback
print("\n========== RESUME FEEDBACK ==========\n")

print("Resume Score:", score)

if len(missing_skills) == 0:
    print("\nExcellent Resume!")
else:
    print("\nMissing Skills:")
    for skill in missing_skills:
        print("-", skill)

# Improvement Tips
print("\n========== IMPROVEMENT TIPS ==========\n")
print("1. Add more technical skills.")
print("2. Include certifications.")
print("3. Add internship experience.")
print("4. Improve communication skills section.")

# Save Feedback Report
feedback_data = {
    "Resume Score": [score],
    "Missing Skills": [", ".join(missing_skills)]
}

feedback_df = pd.DataFrame(feedback_data)

feedback_df.to_csv(
    "resume_feedback_report.csv",
    index=False
)

print("\nFeedback Report Saved Successfully!")

# Conclusion
print("\nProject Completed Successfully!")
```
